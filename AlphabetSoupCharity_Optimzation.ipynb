```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns=["EIN", "NAME"], axis=1)
application_df
```

```python
application_df.nunique()
```

```python
application_type = application_df.APPLICATION_TYPE.value_counts()
application_type
```

```python
replace_application = list(application_type[application_type < 200].index)

for app in replace_application:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
application_df.APPLICATION_TYPE.value_counts()
```

```python
classification = application_df.CLASSIFICATION.value_counts()
classification
```

```python
replace_class = list(classification[classification < 2000].index)

for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")

application_df.CLASSIFICATION.value_counts()
```

```python
app_cat = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']
```

```python
enc = OneHotEncoder(sparse=False)

encode_df = pd.DataFrame(enc.fit_transform(application_df[app_cat]))

encode_df.columns = enc.get_feature_names(app_cat)
encode_df.head()
```

```python
application_df = application_df.merge(encode_df, left_index = True, right_index = True).drop(columns = app_cat, axis = 1)
application_df
```

```python
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop(['IS_SUCCESSFUL'], 1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = len(X_train_scaled[0])
nodes_hidden_layer1 = 8
nodes_hidden_layer2 = 30
nodes_hidden_layer3 = 20

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = nodes_hidden_layer1, activation = "relu", input_dim = num_input_features))
# Second hidden layer
nn.add(tf.keras.layers.Dense(units = nodes_hidden_layer2, activation = "sigmoid"))

# third hidden layer n
nn.add(tf.keras.layers.Dense(units = nodes_hidden_layer3, activation = "sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss = "binary_crossentropy", optimizer="adam", metrics = ["accuracy"] )
```

```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint

os.makedirs("checkpoints/",exist_ok=True)
checkpoint_path = "checkpoints/weights.{epoch:02d}.hdf5"
```

```python

cp_callback = ModelCheckpoint(
    filepath = checkpoint_path,
    verbose = 1,
    save_weights_only = True, 
    save_freq = 'epoch',
    period = 5
    )
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100,callbacks=[cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
